## Bar plot
- bivariate analysis
- numerical vs catagorical
- use case Aggregate  analysis of groups

```python
children=[10,20,30,33,22]
color=['Red','Blue','Green','yellow','pink']
```

```python
import matplotlib.pyplot as plt
plt.bar(color,children,color="Red")
```

```python
plt.barh(color,children)
plt.xlabel('children')
plt.ylabel('color')
```

```python
#mulitple bar charts
import pandas as pd
df=pd.read_csv('dataset/batsman_season_record.csv')
```

```python
df
```

```python
plt.bar(df['batsman'],df['2015'])
plt.bar(df['batsman'],df['2016'])
plt.bar(df['batsman'],df['2017'])
```

```python
import numpy as np
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2,color='blue')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='red')
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2,color='brown')
plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.show()
```

```python
plt.bar(df['batsman'],df['2015'],label='2015')
plt.bar(df['batsman'],df['2016'],bottom=df['2015'],label='2016')
plt.bar(df['batsman'],df['2017'],bottom=df['2015']+df['2016'],label='2017')
plt.legend()

```
