```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('dataset/iris.csv')
```

```python
df
```

```python
plt.scatter(df['SepalLengthCm'],df['PetalLengthCm'])
plt.xlabel('Sepal Length')
plt.ylabel('petal length')
```

```python

df.sample(5)
```

```python
df['Species']= df['Species'].replace({'Iris-setosa':0,'Iris-versicolor':2,'Iris-virginica':3})
```

```python
plt.style.use('fivethirtyeight')
```

```python
df.sample(5)
```

```python
plt.scatter(df['SepalLengthCm'],df['PetalLengthCm'],c=df['Species'],cmap="winter",alpha=0.8)
plt.xlabel('sepal length')
plt.ylabel('petal lenth')
plt.colorbar()
# cmp means color map
```

```python
plt.figure(figsize=(9,5))
plt.scatter(df['SepalLengthCm'],df['PetalLengthCm'],c=df['Species'],cmap="jet",alpha=0.8)
plt.xlabel('sepal length')
plt.ylabel('petal lenth')
plt.colorbar()
```

```python
batters=pd.read_csv('dataset/batter.csv')
```

```python
df=batters.sample(25)
```

```python
df
```

```python
# use fo annotation
plt.figure(figsize=(18,10))
plt.scatter(df['avg'],df['strike_rate'],s=df['runs'])
for i in range(df.shape[0]):
    plt.text(df['avg'].values[i],df['strike_rate'].values[i],df['batter'].values[i])
```

```python
plt.figure(figsize=(18,10))
plt.scatter(df['avg'],df['strike_rate'],s=df['runs'])
for i in range(df.shape[0]):
    plt.text(df['avg'].values[i],df['strike_rate'].values[i],df['batter'].values[i])
plt.axhline(100,color='red')
plt.axvline(25,color='blue')
plt.axhline(140,color='black')
```
