## Subplots

```python
plt.figure(figsize=(15,8))
plt.scatter(batters['avg'],batters['strike_rate'])
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(15,8))
ax.scatter(batters['avg'],batters['strike_rate'],color='red',marker='*')
ax.set_title('subplot example')
ax.set_xlabel('AVg')
ax.set_ylabel('strike reate')
fig.show()
```

```python
plt.style.use('fivethirtyeight')
fig,ax=plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(10,6))
ax[0].scatter(batters['avg'],batters['strike_rate'],color='purple')
ax[1].scatter(batters['avg'],batters['runs'])
ax[0].set_title('AVG vs Strike rate')
ax[0].set_xlabel('AVG')
ax[0].set_ylabel('Strike rate')

ax[1].set_title('avg vs  runs')
ax[1].set_xlabel('AVG')
ax[1].set_ylabel('runs')
```

```python
fix,ax=plt.subplots(nrows=2,ncols=2)
ax[0,0].scatter(batters['avg'],batters['strike_rate'])
ax[0,0].set_title('plot1')

ax[0,1].hist(batters['avg'])
ax[0,1].set_title('plot2')

ax[1,0].scatter(batters['runs'],batters['strike_rate'])
ax[1,1].hist(batters['runs'])
```

```python
fig=plt.figure()
ax1=fig.add_subplot(1,2,1)
ax1.scatter(batters['avg'],batters['strike_rate'],color='red')
ax2=fig.add_subplot(1,2,2)
ax2.hist(batters['runs'])
```

## 3D surface plot


```python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
```

```python
x
```

```python
xx,yy=np.meshgrid(x,y)
xx.shape
```

```python
m=xx**2+yy**2
```

```python
m.shape
```

```python
fig=plt.figure(figsize=(12,8))
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,m,cmap='viridis')
fig.colorbar(p)
```

```python

z=np.sin(xx)+np.cos(yy)
fig=plt.figure(figsize=(12,8))
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,m,cmap='viridis')
fig.colorbar(p)

```

```python
fig=plt.figure(figsize=(12,8))
ax=plt.subplot()
p=ax.contour(xx,yy,m,cmap='viridis')
fig.colorbar(p)
```

```python
fig=plt.figure(figsize=(12,8))
ax=plt.subplot()
p=ax.contourf(xx,yy,m,cmap='viridis')
fig.colorbar(p)
```
