## Scatter plot
- bivariate
- num vs num
- use case :finding correlation 
- it is like 2d plot 

```python
import pandas as pd
import numpy as np
import random
x=np.linspace(-10,10,50)
y=10*x+3+ np.random.randint(0,300,50)
```

```python
x
```

```python
y
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
#plot scatter plot using dataset
df=pd.read_csv('dataset/batter.csv')
```

```python
df
```

```python
df=df.head(50)
```

```python
plt.scatter(df['avg'],df['strike_rate'],color='red',marker="*")
plt.title('AVg and strike rate analysis of top 50 player')
plt.xlabel('average')
plt.ylabel('strive rate')
plt.show()
```

```python
import seaborn as sns
tips=sns.load_dataset('tips')
tips
```

```python
plt.scatter(tips['total_bill'],tips['tip'],s=tips['size']*20)
```
