## Basic graph


```python
x=[1,2,3]
y=[2,3,4]
plt.plot(x,y,label='2X',color='red')
plt.title("first graph")
plt.xlabel('x')
plt.ylabel('y')
plt.xticks([1,2,3,4])
plt.yticks([2,4,6,8])
plt.legend()
plt.show()
```

```python
# from pandas dataset
batsman=pd.read_csv('dataset/sharma-kohli.csv')
```

```python
batsman
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
#plot multiple plots
plt.plot(batsman['index'],batsman['RG Sharma'],color='green',linestyle='dashed')
plt.plot(batsman['index'],batsman['V Kohli'],color='blue',linestyle='dashed')
plt.title('Rohit VS Virat kohli carier comparision')
plt.xlabel('season')

plt.ylabel('run score')
plt.show()
```

```python
#plot multiple plots
plt.plot(batsman['index'],batsman['RG Sharma'],color='green',linestyle='dashdot',linewidth=3)
plt.plot(batsman['index'],batsman['V Kohli'],color='blue',linestyle='dashdot',linewidth=2)
plt.title('Rohit VS Virat kohli carier comparision')
plt.xlabel('season')

plt.ylabel('run score')
plt.show()
```

```python
#plot multiple plots
plt.plot(batsman['index'],batsman['RG Sharma'],color='green',linestyle='dashed',marker='d',markersize=10)
plt.plot(batsman['index'],batsman['V Kohli'],color='blue',linestyle='dashed',marker='*')
plt.title('Rohit VS Virat kohli carier comparision')
plt.xlabel('season')

plt.ylabel('run score')
plt.show()
```

```python
#legend and location
plt.plot(batsman['index'],batsman['RG Sharma'],color='green',linestyle='dashed',marker='d',markersize=10,label='Rohit')
plt.plot(batsman['index'],batsman['V Kohli'],color='blue',linestyle='dashed',marker='*',label='Virat')
plt.title('Rohit VS Virat kohli carier comparision')
plt.xlabel('season')
plt.ylabel('run score')
plt.legend()
plt.grid()
plt.show()
```
