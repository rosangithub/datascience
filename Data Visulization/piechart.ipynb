## Pie charts
- univariate/bivariate
- catagorical vs numerical
- use case :to find contribution on standard scale


```python
from matplotlib import pyplot as plt
data=[23,45,100,20,49]
subjects=['eng','math','science','physics','nepali']

plt.pie(data,labels=subjects)
plt.show()
```

```python
import pandas as pd
df=pd.read_csv('dataset/gayle-175.csv')
```

```python
df
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
plt.legend()
plt.show()
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','red','black'])
plt.legend()
plt.show()
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.1,0,0,0.3,0,0.1],shadow=True)
plt.legend()
plt.show()
```
