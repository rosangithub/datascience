```python
import pandas as pd
import numpy as np 
```

```python
# These are the headers which you need to link to your dataset.
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df=pd.read_csv("../youtube/dataset/car_dataset.data",names=headers)
```

```python
df
```

```python
df.head()
```

```python
df.isnull()
```

```python
df.replace("?",np.nan,inplace=True)
```

```python
df.head()
```

```python
df.isnull()
```

```python
missing_data=df.isnull()
missing_data.head()
```

```python
for column in missing_data.columns.values.tolist():
    print(missing_data[column].value_counts())
```

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
ls=['normalized-losses','stroke','bore','horsepower','peak-rpm']

imputer.fit(df[ls])
df[ls]=imputer.transform(df[ls])
```

```python
df.head()
```

```python
imputer=SimpleImputer(missing_values=np.nan,strategy='most_frequent')
imputer.fit(df[['num-of-doors']])
df[['num-of-doors']]=imputer.transform(df[['num-of-doors']])
```

```python
df.dropna(subset=['price'],axis=0,inplace=True)
df.reset_index(drop=True,inplace=True)
```

```python
missing_data=df.isnull()
```

```python
missing_data.head()
```

```python
for column in missing_data.columns.values.tolist():
    print(missing_data[column].value_counts())
```

```python
df.info()
```

```python
bins=np.linspace(min(df['horsepower']),max(df['horsepower']),4)
```

```python
group_names=['low','medium','high']
```

```python
df['horsepower_bins']=pd.cut(df['horsepower'],bins,labels=group_names,include_lowest=True)
df[['horsepower','horsepower_bins']].head(20)
```

```python
df['horsepower_bins'].value_counts()
```

```python
plt.bar(group_names,df['horsepower_bins'].value_counts())
plt.xlabel('horsepower')
plt.ylabel('counts')
```
